```python
"""
Maximum subarray
"""


def max_subarray(nums):
    dp = [0] * (len(nums) + 1)

    for i in range(len(nums)):
        dp[i] = max(nums[i], dp[i - 1] + nums[i])

    return max(dp)
```

```python
"""
Gas Station
"""


def gas_station(gas, cost):
    #If the sum of all the available gas is less than the sum of the costing gas
    if sum(gas) < sum(cost):
        return -1

    total = 0
    res = 0

    for i in range(len(gas)):
        total += (gas[i] - cost[i])

        if total < 0:
            total = 0
            res = i + 1

    return res


def gas_station(gas, cost):
    if sum(gas) < sum(cost):
        return -1

    res = 0
    total = 0

    for i in range(len(gas)):
        total += (gas[i] - cost[i])

        if total < 0:
            total = 0
            res += 1

    return res
```

```python
"""
Hand of straights
"""

from collections import Counter
import heapq


def hand_of_straights(cards, groupsize):
    #The count of all cards must be a multiple of the group size, otherswise no solution is possible
    if len(cards) % groupsize:
        return False

    #Give the count of all the cards in the cards array
    count = Counter(cards)

    #Create a min_heap of all the unique cards in the count dictionary
    min_heap = heapq.heapify(list(count.keys()))

    #While the min heap has elements
    while min_heap:
        #Pick the first element of the min heap, which is the least
        first = min_heap[0]

        #Iterate through the range of the value of count to the value of count + groupsize as the cards need to be consecutive
        for i in range(first, first + groupsize):
            #If the card is no longer in count, then the solution is not possible
            if i not in count:
                return False

            #Decrement the value of card in count map
            count[i] -= 1
            #If the count map value is zero
            if count[i] == 0:
                #It should be the least element in the min heap, otherwise return false
                if i != min_heap[0]:
                    return False
                #Pop this element from the min heap
                heapq.heappop(min_heap)

    return True


"""
Hand of straights again
Card array must be divided into consecutive piles of cards of group size, group_size, return True if possible, else return False
"""
from collections import Counter
import heapq


def hand_of_straights(cards, group_size):
    if len(cards) % group_size != 0:
        return False

    count = Counter(cards)

    min_heap = heapq.heapify(list(count.keys()))

    while min_heap:
        first = min_heap[0]

        for i in range(first, first + group_size):
            if i not in count:
                return False

            count[i] -= 1

            if count[i] == 0:
                if i == min_heap[0]:
                    return False
                heapq.heappop(min_heap)

    return True
```

```python
"""
Merge triplets to form target triplet
"""


def merge_triplets(triplets, target):
    good = set()

    for t in triplets:
        if t[0] > target[0] or t[1] > target[1] or t[2] > target[2]:
            continue

        for i, v in enumerate(t):
            if v == target[i]:
                good.add(i)

    return len(good) == 3


def merge_triplets(triplets, target):
    good = set()

    for t in triplets:
        if t[0] > target[0] or t[1] > target[1] or t[2] > target[2]:
            continue

        for i, v in enumerate(t):
            if v == target[i]:
                good.add(i)

    return len(good) == 3
```

```python
"""
Parition labels
"""
from collections import defaultdict


def partition_labels(sr):
    lastIndex = defaultdict(int)

    for i, c in enumerate(sr):
        lastIndex[c] = i

    res = []
    size = end = 0

    for i, c in enumerate(sr):
        size += 1
        end = max(end, lastIndex[c])

        if i == end:
            res.append(size)
            size = 0

    return res
```

```python
"""
Valid parenthesis
"""


def check_valid_string(sr):
    left = []
    star = []

    for i, ch in enumerate(sr):
        if ch == '(':
            left.append(i)
        elif ch == "*":
            star.append(i)

        else:
            if not left and not star:
                return False

            if left:
                left.pop()

            else:
                star.pop()

    while left and star:
        if left.pop() > star.pop():
            return False

    return not left
```
