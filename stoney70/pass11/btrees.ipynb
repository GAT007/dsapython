```python
"""
Average of levels in binary tree
"""

from collections import deque


def average_of_levels(root):
    if not root:
        return []

    avgs = []
    q = deque([root])

    while q:
        temp = []
        for i in range(len(q)):
            curr = q.popleft()
            temp.append(curr)
            if curr.left: q.append(curr.left)
            if curr.right: q.append(curr.right)
        avgs.append(sum(temp) / len(temp))

    return avgs
```

```python
"""
Breadth First Search
"""

from collections import deque


def bfs(root):
    if not root:
        return None

    q = deque([root])
    res = []

    while q:
        tmp = []
        for i in range(len(q)):
            curr = q.popleft()
            tmp.append(curr.val)
            if curr.left: q.append(curr.left)
            if curr.right: q.append(curr.right)

        res.append(tmp)

    return res
```

```python
"""
Depth First Search
"""

"""
Preorder iterative
"""


def preorder_iterative(root):
    if not root:
        return None

    stack = [root]
    res = []

    while stack:
        curr = stack.pop()
        res.append(curr)
        if curr.left: stack.append(curr.left)
        if curr.right: stack.append(curr.right)

    return res


def preorder_recursive(root):
    if not root:
        return None

    res = []

    def dfs(curr):
        res.append(curr)
        if curr.left: dfs(curr.left)
        if curr.right: dfs(curr.right)

    dfs(root)

    return res


"""
Inorder iterative
"""


def inorder_iterative(root):
    if not root:
        return []

    stack = []
    res = []
    curr = root

    while curr or stack:
        while curr:
            stack.append(curr)
            curr = curr.left

        curr = stack.pop()
        res.append(curr.val)

        curr = curr.right

    return res


def inorder_recursive(root):
    if not root:
        return []

    res = []

    def dfs(curr):
        if not curr:
            return None

        if curr.left: dfs(curr.left)

        res.append(curr.val)

        if curr.right: dfs(curr.right)

    return res


"""
Postorder traversal
"""


def postorder_iterative(root):
    if not root:
        return []

    stack = [root]
    res = []

    while stack:
        curr = stack.pop()
        res.append(curr.val)

        if curr.left: res.append(curr.left)
        if curr.right: res.append(curr.right)

    return res[::-1]


def postorder_recursive(root):
    if not root:
        return []

    res = []

    def dfs(curr):
        if curr.left: dfs(curr.left)
        if curr.right: dfs(curr.right)

        res.append(curr.val)

    dfs(root)

    return res
```

```python
"""
Minimum/Maximum depth of a binary tree
"""

from collections import deque


def min_and_max_depth(root):
    if not root:
        return None

    levels = 0
    min_depth = float('inf')
    flag = False
    q = deque([root])

    while q:
        levels += 1
        for i in range(len(q)):
            curr = q.popleft()
            if not curr.left and not curr.right and not flag:
                min_depth = min(levels, min_depth)
                flag = True
            if curr.left: q.append(curr.left)
            if curr.right: q.append(curr.right)

    return min_depth, levels
```

```python
"""
Min/Max Value of a binary tree
"""
from collections import deque


def min_max_values(root):
    if not root:
        return None

    min_value = float('inf')
    max_value = float('-inf')

    q = deque([root])

    while q:
        curr = q.popleft()
        min_value = min(min_value, curr.val)
        max_value = max(max_value, curr.val)
        if curr.left: q.append(curr.left)
        if curr.right: q.append(curr.right)

    return min_value, max_value
```

```python
"""
Binary Tree Level Order Traversal
"""

from collections import deque


def level_order_traversal(root):
    if not root:
        return []

    q = deque([root])
    avgs = []

    while q:
        temp = []
        for i in range(len(q)):
            curr = q.popleft()
            temp.append(curr.val)
            if curr.left: q.append(curr.left)
            if curr.right: q.append(curr.right)
        avgs.append(temp)

    return avgs
```

```python
"""
Same Tree
"""


def same_tree(p, q):
    if not p and not q:
        return True

    if not p or not q:
        return False

    stack = [(p, q)]
    while stack:

        r1, r2 = stack.pop()

        if not r1 and not r2:
            continue

        if not r1 or not r2:
            return False

        if r1.val != r2.val:
            return False

        stack.append((r1.right, r2.right))
        stack.append((r1.left, r2.left))

    return True
```

```python
"""
Path Sum
"""
from collections import deque


def path_sum(root, target):
    q = deque([(root, root.val)])

    while q:
        curr, value = q.popleft()
        if value == target and not curr.left and not curr.right:
            return True

        if curr.left:
            q.append((curr.left, value + curr.left.val))

        if curr.right:
            q.append((curr.right, value + curr.right.val))

    return False
```

```python
"""
Diameter of a binary tree

TC O(n)
"""


def diameter(root):
    if not root:
        return 0

    diameter = 0

    def depth(curr):
        if not curr:
            return 0
        nonlocal diameter

        left = depth(curr.left)
        right = depth(curr.right)

        diameter = max(diameter, left + right)
        return 1 + max(left, right)

    depth(root)

    return diameter
```

```python
"""
Invert Binary Tree
"""


def invert_binary_tree(root):
    if not root:
        return None

    stack = [root]

    while stack:
        curr = stack.pop()
        curr.left, curr.right = curr.right, curr.left
        if curr.right: stack.append(curr.right)
        if curr.left: stack.append(curr.left)

    return root
```

```python
"""
Mirror of binary tree
"""


def mirror(p, q):
    def find_mirror(r1, r2):
        if not r1 and not r2:
            return True

        if not r1 or not r2:
            return False

        if r1.val != r2.val:
            return False

        return (r1.val == r2.val) and (find_mirror(r1.left, r2.right)) and (find_mirror(r1.right, r2.left))

    return find_mirror(p, q)
```

```python
"""
LCA of a binary tree
"""

from collections import deque


def lca(root, p, q):
    if not root or not p or not q:
        return None

    parent_map = {root: None}

    qu = deque([root])

    while qu:
        curr = qu.popleft()
        if p and q in parent_map:
            break

        if curr.left:
            parent_map[curr.left] = curr
            qu.append(curr.left)

        if curr.right:
            parent_map[curr.right] = curr
            qu.append(curr.right)

    ancestors = set()

    while p:
        ancestors.add(p)
        p = parent_map[p]

    while q:
        if q in ancestors:
            return q
        q = parent_map[q]

    return False
```

```python
"""
Balanced Tree
"""


def isbalanced(root):
    if not root:
        return True

    def dfs(curr):
        if not curr:
            return [True, 0]

        left = dfs(curr.left)
        right = dfs(curr.right)

        balanced = left[0] and right[0] and abs(left[1] - right[1]) <= 1
        return [balanced, 1 + max(left[1], right[1])]

    return dfs(root)[0]
```
